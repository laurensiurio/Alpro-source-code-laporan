```python
# capitalize() - Mengubah huruf pertama menjadi huruf kapital
kata = "indonesia"
print(kata.capitalize())  # Output: Indonesia

# count() - Menghitung berapa kali suatu substring muncul dalam string
print(kata.count('i'))  # Output: 2

# endswith() - Memeriksa apakah string diakhiri dengan substring tertentu
print(kata.endswith('a'))  # Output: True

# startswith() - Memeriksa apakah string dimulai dengan substring tertentu
print(kata.startswith('in'))  # Output: True

# find() - Mencari posisi pertama dari substring dalam string
print(kata.find('do'))  # Output: 5

# islower() - Memeriksa apakah semua karakter dalam string adalah huruf kecil
print(kata.islower())  # Output: True

# isupper() - Memeriksa apakah semua karakter dalam string adalah huruf besar
print(kata.isupper())  # Output: False

# isdigit() - Memeriksa apakah semua karakter dalam string adalah angka
print(kata.isdigit())  # Output: False

# strip() - Menghapus whitespace di awal dan akhir string
kata_spasi = "  indonesia  "
print(kata_spasi.strip())  # Output: indonesia

# split() - Memisahkan string menjadi list berdasarkan separator
kalimat = "saya belajar python"
print(kalimat.split())  # Output: ['saya', 'belajar', 'python']
```
